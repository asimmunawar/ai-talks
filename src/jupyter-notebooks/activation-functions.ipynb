Activation functions play a crucial role in neural networks by introducing non-linearity to the model, allowing it to learn complex relationships in the data. Here are some commonly used activation functions, along with explanations and Python code snippets to plot them:

## Sigmoid Activation Function:

Formula: 

$f(x) = \frac{1}{1 + e^{-x}}$

Explanation: Squeezes the output between 0 and 1, often used in the output layer for binary classification problems.

Python code snippet to plot:

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

## Sigmoid with its derivative

$\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))$

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

x = np.linspace(-10, 10, 500)  # Generate 500 points from -10 to 10

sigmoid_values = sigmoid(x)
sigmoid_derivative_values = sigmoid_derivative(x)

plt.figure(figsize=(10, 6))

plt.plot(x, sigmoid_values, label='Sigmoid', color='blue', linewidth=2)
plt.plot(x, sigmoid_derivative_values, label='Derivative of Sigmoid', color='red', linestyle='dotted')

plt.axvline(x=0, color='black', linestyle='--', label='y-axis')

plt.title('Sigmoid Activation Function and Its Derivative', fontsize=16)
plt.xlabel('x', fontsize=18)
plt.ylabel('Value', fontsize=18)
plt.ylim(0, 1)
plt.legend(fontsize=18)

plt.tight_layout()
plt.show()
```

## ReLU (Rectified Linear Activation) Function:

Formula: 

$f(x) = max(0, x)$

Explanation: Outputs the input directly if it's positive, otherwise, it outputs zero. This helps with mitigating the vanishing gradient problem.

Python code snippet to plot:

```python
def relu(x):
    return np.maximum(0, x)

y = relu(x)

plt.plot(x, y)
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

## Leaky ReLU Function:


Formula: 
$f(x) = \left\{\begin{matrix}x, \textup{  if } x> 0\\ \alpha x, \textup{ 
  otherwise}\end{matrix}\right.$

Here, $\alpha$ is a small positive constant, often around 0.01.

Explanation: Addresses the "dying ReLU" problem by allowing a small gradient for negative inputs, preventing neurons from becoming inactive.

Python code snippet to plot:

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

y = leaky_relu(x)

plt.plot(x, y)
plt.title("Leaky ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

## Tanh (Hyperbolic Tangent) Function:

Formula: 
$f(x) = tanh(x)$

$f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$
 

Explanation: Similar to the sigmoid function, but outputs values in the range 
1 to 1, centered around 0.

Python code snippet to plot:


```python
def tanh(x):
    return np.tanh(x)

y = tanh(x)

plt.plot(x, y)
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```
